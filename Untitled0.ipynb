19. Створіть змінну centered_price, яка містить версію стовпця price з відніманням середньої ціни.

(Примітка: це «центрувальне» перетворення є поширеним кроком попередньої обробки перед застосуванням різних алгоритмів машинного навчання.)

```python
# 1. Обчислюємо середню ціну (mean)
mean_price = reviews['price'].mean()

# 2. Створюємо змінну centered_price, віднімаючи середнє від стовпця price
centered_price = reviews['price'] - mean_price

# Перевірка результату: виведемо перші 5 значень
print(f"Середня ціна по всьому датасету: {mean_price:.2f}")
print("\nЦентровані ціни (перші 5):")
print(centered_price.head())
```

20. Я економний покупець вина. Яке вино є "найкращою вигідною покупкою"? Створіть змінну bargain_wine з назвою вина з найвищим співвідношенням балів до ціни в наборі даних.

```python
# 1. Створюємо тимчасову серію з відношенням балів до ціни
# Ми ділимо 'points' на 'price'
points_per_price = reviews['points'] / reviews['price']

# 2. Знаходимо індекс рядка, де це співвідношення є максимальним
# idxmax() повертає індекс (назву рядка) з найбільшим значенням
best_bargain_index = points_per_price.idxmax()

# 3. Витягуємо назву вина (title) за цим індексом
bargain_wine = reviews.loc[best_bargain_index, 'title']

# Виводимо результат
print(f"Найвигідніше вино: {bargain_wine}")
print(f"Показник вигідності: {points_per_price.max():.4f} балів за 1$")
```

22. Ми хотіли б розмістити ці огляди вин на нашому вебсайті, але система оцінювання від 80 до 100 балів занадто складна для розуміння – ми хотіли б перевести їх у прості зіркові рейтинги. Оцінка 95 або вище вважається 3 зірками, оцінка щонайменше 85, але менше 95 – 2 зірками. Будь-яка інша оцінка – 1 зірка.

Крім того, Канадська асоціація виноробів придбала багато реклами на сайті, тому будь-які вина з Канади повинні автоматично отримувати 3 зірки, незалежно від балів.

Створіть серію star_ratings (тип змінної Series) з кількістю зірок, що відповідає кожному огляду в наборі даних.

```python
import pandas as pd

# 1. Визначаємо функцію для перетворення балів та країни у зірки
def convert_to_stars(row):
    # Умова для Канади (пріоритетна)
    if row['country'] == 'Canada':
        return 3
    # Умова для високих балів
    elif row['points'] >= 95:
        return 3
    # Умова для середніх балів
    elif row['points'] >= 85:
        return 2
    # Всі інші випадки
    else:
        return 1

# 2. Застосовуємо функцію до кожного рядка датафрейму
# axis='columns' (або 1) дозволяє функції бачити весь рядок, а не окремі колонки
star_ratings = reviews.apply(convert_to_stars, axis='columns')

# 3. Перевірка результату
print("Розподіл зіркових рейтингів:")
print(star_ratings.value_counts())

# Виведемо кілька прикладів для перевірки (країна, бали, зірки)
test_check = pd.concat([reviews[['country', 'points']], star_ratings.rename('stars')], axis=1)
print("\nПриклади розрахунку:")
print(test_check.head())
```

25. Які мінімальні та максимальні ціни для кожного сорту (variety) вина? Створіть DataFrame, індекс якого є категорією сорту з набору даних, а значення якого – її мінімальним та максимальним значеннями.

```python
# 1. Групуємо дані за сортом вина ('variety')
# 2. Вибираємо колонку ціни ('price')
# 3. Застосовуємо агрегацію для пошуку min та max одночасно
price_extremes = reviews.groupby('variety')['price'].agg(['min', 'max'])

# 4. Виведемо результат для перших 5 сортів
print("Мінімальні та максимальні ціни за сортами вина:")
print(price_extremes.head())

# Перевірка: сортування за максимальною ціною, щоб побачити найдорожчі сорти
print("\nТоп-5 найдорожчих сортів за максимальною ціною:")
print(price_extremes.sort_values(by='max', ascending=False).head())
```

31. Іноді стовпець ціни не містить значення. Скільки відгуків у наборі даних не мають ціни?

```python
# 1. Знаходимо всі порожні значення в стовпці 'price'
# .isnull() повертає True для кожної порожньої клітинки
missing_prices = reviews['price'].isnull()

# 2. Рахуємо суму (True рахується як 1, False як 0)
n_missing_prices = missing_prices.sum()

# Виводимо результат
print(f"Кількість відгуків без вказаної ціни: {n_missing_prices}")

# Додатково: відсоток від загальної кількості даних
percent_missing = (n_missing_prices / len(reviews)) * 100
print(f"Це становить приблизно {percent_missing:.2f}% від усього набору даних.")
```

Частина 2. Matplotlib


На оцінку задовільно: створити одновимірні графіки на основі датасету із частини 1 усіх типів:



```python
import pandas as pd
import matplotlib.pyplot as plt

# Налаштування розміру графіків для зручності
plt.rcParams['figure.figsize'] = (10, 5)

# 1. Bar Chart (Стовпчикова діаграма)
# Найкраще підходить для категоріальних даних (наприклад, топ-10 країн)
reviews['country'].value_counts().head(10).plot.bar()
plt.title("Топ-10 країн за кількістю вин")
plt.show()

# 2. Line Chart (Лінійний графік)
# Добре показує тенденції (наприклад, як змінюється кількість балів)
# Сортуємо за індексом, щоб лінія була логічною
reviews['points'].value_counts().sort_index().plot.line()
plt.title("Розподіл оцінок (лінійний графік)")
plt.show()

# 3. Area Chart (Діаграма з областями)
# Схожа на лінійну, але заповнює простір під лінією
reviews['points'].value_counts().sort_index().plot.area()
plt.title("Розподіл оцінок (діаграма з областями)")
plt.show()

# 4. Histogram (Гістограма)
# Ідеально для інтервальних даних, наприклад, розкид цін
# Використовуємо обмеження [price < 100], щоб графік був розбірливим
reviews[reviews['price'] < 100]['price'].plot.hist(bins=20)
plt.title("Гістограма цін на вина (до $100)")
plt.show()
```

На оцінку добре (додатково до завдання на оцінку задовільно): створити багатовимірні графіки на основі датасету із частини 1 усіх типів:

```python
# Вибираємо вина дешевше $100, щоб графік не був занадто розтягнутим
reviews[reviews['price'] < 100].sample(1000).plot.scatter(x='price', y='points')
plt.title("Взаємозв'язок ціни та балів (Scatter Plot)")
plt.show()

reviews[reviews['price'] < 100].plot.hexbin(x='price', y='points', gridsize=15)
plt.title("Щільність розподілу ціна/бали (Hex Plot)")
plt.show()

# Створюємо зведену таблицю для 5 найпопулярніших сортів
top_varieties = reviews[reviews['variety'].isin(reviews['variety'].value_counts().head(5).index)]
variant_counts = top_varieties.groupby(['points', 'variety']).size().unstack().fillna(0)

variant_counts.plot.bar(stacked=True)
plt.title("Кількість оцінок за сортами вин (Stacked Bar)")
plt.show()

variant_counts.plot.line()
plt.title("Тенденції оцінок для різних сортів (Bivariate Line)")
plt.show()
```

На оцінку відмінно (додатково до завданнь на оцінку задовільно та добре): побудувати графіки кореляції змінних на основі датасету із частини 1 усіх типів:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Беремо 3 популярних сорти для чіткості
top_3 = reviews[reviews['variety'].isin(['Pinot Noir', 'Chardonnay', 'Cabernet Sauvignon'])].sample(300)

sns.scatterplot(x='price', y='points', hue='variety', data=top_3)
plt.title("Ціна vs Бали з розбивкою за сортами")
plt.show()

# Вибираємо топ-5 сортів
top_5_varieties = reviews[reviews.variety.isin(reviews.variety.value_counts().head(5).index)]

sns.boxplot(x='variety', y='points', data=top_5_varieties)
plt.title("Розподіл балів за популярними сортами")
plt.xticks(rotation=45)
plt.show()

# Обчислюємо матрицю кореляції для числових даних
corr_matrix = reviews[['points', 'price']].corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Теплокарта кореляції ціни та балів")
plt.show()

from pandas.plotting import parallel_coordinates

# Готуємо дані: вибираємо 2 сорти та потрібні колонки
cols = ['points', 'price', 'variety']
df_parallel = top_5_varieties[cols].sample(50).dropna()

# Масштабуємо ціну для візуалізації поруч із балами (від 0 до 1)
df_parallel['price'] = (df_parallel['price'] - df_parallel['price'].min()) / (df_parallel['price'].max() - df_parallel['price'].min())
df_parallel['points'] = (df_parallel['points'] - df_parallel['points'].min()) / (df_parallel['points'].max() - df_parallel['points'].min())

parallel_coordinates(df_parallel, 'variety', color=['#FF5733', '#33FF57', '#3357FF', '#F333FF', '#FFB833'])
plt.title("Профілі вин у паралельних координатах")
plt.show()
```
